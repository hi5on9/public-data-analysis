```python
import os
import sys
import urllib.request
import datetime
import json
import xmltodict
import math
import requests
from bs4 import BeautifulSoup
```

```python
# api 접근 
url = 'http://data.ex.co.kr/openapi/locationinfo/locationinfoIc' 
parameters = {"key":"1190242257", "type":"xml"} 
req = requests.get(url, params=parameters) 
```

```python
html = req.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
total = []

for i in range(len(soup.find_all('iccode'))) :
    iccode = soup.find_all('iccode')[i].text
    icname = soup.find_all('icname')[i].text
    routename = soup.find_all('routename')[i].text
    routeno = soup.find_all('routeno')[i].text
    xvalue = soup.find_all('xvalue')[i].text
    yvalue = soup.find_all('yvalue')[i].text
    total.append([iccode]+[icname]+[routename]+[routeno]+[xvalue]+[yvalue])
```

```python
total
```

```python
import pandas as pd
df = pd.DataFrame(total,columns=(['iccode','icname','routename','routeno','xvalue','yvalue']))
df                              
```
